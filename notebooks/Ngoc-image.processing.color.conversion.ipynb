# Lý thuyết không gian màu

## Không gian màu là gì ?
- Không gian màu (color space hoặc color system) là một không gian hay một mô hình giúp máy tính hoặc con người xác định được giá trị màu thông qua các phương pháp hoặc công thức cụ thể. Bình thường hiện nay thì không gian màu thường được biểu diễn qua hệ thống tọa độ và mỗi màu được biểu diễn là một điểm trong không gian đó (Digital Image Processing Edition 4th)

## Các không gian màu phổ biến
- **Không gian màu Grayscale**: Mỗi một pixel được biểu diễn bởi một giá trị mức xám trong khoảng [0, 255]
- **Không gian màu RGB**: Biểu diễn thông qua ba kênh màu cơ bản gồm đỏ (Red), xanh lá (Green) và xanh dương (Blue). Giá trị của các kênh màu trải dài từ [0, 255]
- **Không gian màu HSV**: Biểu diễn các màu thông qua phối hợp các thông tin về 
  - Màu sắc (Hue): Biểu diễn giá trị màu sắc trong vòng quang phổ với đơn vị góc có giá trị [0, 179]
  - Độ bão hòa màu (Saturation): Biểu diễn mức độ bão hòa với mức độ thuộc khoảng [0, 255]
  - Ánh sáng (Value hoặc Bright): Biểu diễn độ sáng của màu thuộc khoảng [0, 255] 
- **Không gian màu CMY/CMYK**: Trong lĩnh vực in ấn, do mực có khả năng hấp thụ ánh sáng từ ánh sáng trắng và phản xạ các màu còn lại. Vì vậy, người ta đã xây dựng không gian màu riêng xác định màu dựa trên mức độ hấp thụ ánh sáng. Không gian màu được biểu diễn bởi ba giá trị
  - Cyan (C): Lượng ánh sáng màu đỏ bị hấp thụ, tính bằng công thức C = 1 - R
  - Magenta (M): Lượng ánh sáng màu xanh lá bị hấp thụ, tính bằng công thức M = 1 - G
  - Yellow (Y): Lượng ánh sáng màu xanh dương bị hấp thụ, tính bằng công thức Y = 1 - B
  - Black (K): Đại lượng biểu thị màu đen do C-M-Y không thể hiện được màu đen thuần.

- **Không gian màu LAB**: Hệ màu được xây dựng trên thị giác của người. Các giá trị mà LAB mô tả tất cả các giá trị mà mắt người có thể nhìn thấy được và độc lập với thiết bị, tức là không phụ thuộc thuộc loại thiết bị, tuổi thọ thiết bị,... mà giới hạn khả năng biểu diễn. Trong không gian có các đại lượng
  - $\Delta L$: Giá trị lớn thì nghiêng về màu trắng và ngược lại nghiêng về màu đen
  - $\Delta a$: Giá trị lớn thì nghiêng về màu đỏ và ngược lại nghiêng về màu xanh lá
  - $\Delta b$: Giá trị lớn thì nghiêng về màu vàng và ngược lại nghiêng về màu xanh dương

## Không gian màu là gì ?
- Không gian màu (color space hoặc color system) theo định nghĩa trong sách Gonzalet là một không gian hay một mô hình giúp máy tính hoặc con người xác định được giá trị màu thông qua các phương pháp hoặc công thức cụ thể. Bình thường hiện nay thì không gian màu thường được biểu diễn qua hệ thống tọa độ và mỗi màu được biểu diễn là một điểm trong không gian đó.

- **Không gian màu CMY/CMYK**: Trong lĩnh vực in ấn, do mực có khả năng hấp thụ ánh sáng từ ánh sáng trắng và phản xạ các màu còn lại. Vì vậy, người ta đã xây dựng không gian màu riêng xác định màu dựa trên mức độ hấp thụ ánh sáng. Không gian màu được biểu diễn bởi ba giá trị
  - Cyan (C): Lượng ánh sáng màu đỏ bị hấp thụ, tính bằng công thức C = 1 - R
  - Magenta (M): Lượng ánh sáng màu xanh lá bị hấp thụ, tính bằng công thức M = 1 - G
  - Yellow (Y): Lượng ánh sáng màu xanh dương bị hấp thụ, tính bằng công thức Y = 1 - B
  - Black (K): Đại lượng biểu thị màu đen do C-M-Y không thể hiện được màu đen thuần.

## Ứng dụng của các không gian màu 
Hiện nay, các không gian màu khác nhau được sử dụng với đa dạng mục đích và tác vụ khác nhau, nhưng nhìn chung đây là một số ứng dụng chính trong việc nghiên cứu và phân tích không gian màu.

- **Grayscale**: Dùng nhiều trong nhận dạng kí tự, camera an ninh do sự đơn giản tính toán, chạy nhanh hơn và tiết kiệm tài nguyên hơn.
- **RGB**: Được sử dụng rộng rãi trong các màn hình điện thoại, TV có độ phân giải cao, đồ họa máy tính, thiết kế Web do khả năng biểu diễn được không gian màu rộng lớn.
- **HSV**: Được sử dụng nhiều trong photoshop, tạo bộ chọn màu,... do không gian màu vừa đủ và gần với mắt người nhận diện, dễ thao tác hơn RGB 
- **LAB**: Được sử dụng trong cách phần mềm chỉnh ảnh chuyên nghiệp như ADOBE, đo độ khác biệt màu do tính nhất quán của không gian, không phụ thuộc vào thiết bị và thống nhất trên nhiều hệ thống.
- **CMY/CMYK**: Được sử dụng rộng rãi trong in ấn, thiết kế màu poster, xuất file in ấn do mục đích kiểm soát màu của tài liệu khi bị hấp thụ bởi ánh sáng. Đặc biệt, CMY còn được ứng dụng trong việc tạo màu cho các đèn sân khấu hiện đại, sử dụng rộng rãi trong các sự kiện, hội nghị và biểu diễn nghệ thuật do các ưu điểm như độ chính xác màu cao, chuyển màu mượt mà, tiết kiệm năng lượng và phù hợp với các buổi diễn nghệ thuật đòi hỏi đèn phù hợp với sân khấu

### Không gian RGB sang không gian Grayscale
Ảnh RGB (3 kênh) khi chuyển sang ảnh xám (1 kênh) thì có công thức tổng hợp trọng số Luminance [1] như sau
$$
G(x,y) = 0.299 \cdot R(x,y) + 0.587 \cdot G(x,y) + 0.114 \cdot B(x,y)
$$

Nhưng từ ảnh Grayscale thì không thể chuyển thành ảnh RGB
### Không gian RGB sang không gian CMY/CMYK
- Để chuyển ảnh RGB sang CMY thì ta có công thức sau (mặc định đã chuẩn hóa ảnh RGB trong đoạn [0,1])
$$
\begin{bmatrix}
C \\
M \\
Y
\end{bmatrix}
=
\begin{bmatrix}
1 \\
1 \\
1
\end{bmatrix}
-
\begin{bmatrix}
R \\
G \\
B
\end{bmatrix}
$$
Hiểu đơn giản là CMY chính là phần bị ánh sáng hấp thụ, phần còn lại phản xạ lại vào mắt, ta có thể viết lại công thức cho hình dung dễ hơn

$$
\begin{bmatrix}
R \\
G \\
B
\end{bmatrix}
=
\begin{bmatrix}
1 \\
1 \\
1
\end{bmatrix}
-
\begin{bmatrix}
C \\
M \\
Y
\end{bmatrix}
$$

Ví dụ nói về khả năng hấp thụ ánh sáng đỏ của giá trị Cyan (C), tương tự như M và Y
|Giá trị C| Kết quả|
|---|---|
| C = 0|Màu đỏ không bị hấp thụ, phản xạ lại hết, người có thể nhìn rõ màu đỏ|
| 0 < C < 1| Màu đỏ bị hấp thụ một phần và phản xạ một phần, người thấy màu đỏ mờ mờ|
| C = 1| Màu đỏ bị hấp thụ hoàn toàn, mắt người không thể nhìn màu đỏ|

Từ công thức trên, hai không gian màu RGB và CMY chuyển qua lại khá đơn giản bằng một phép tính, sau này khi người ta mong muốn tài liêu, poster, quảng cáo phản xạ màu mong muốn thì người ta sẽ chỉnh C,M,Y để hấp thụ các ánh sáng không cần thiết và tạo màu đẹp hơn.
- Chuyển từ không gian CMY sang CMYK thì chỉ cần chọn thêm K = min(C,M,Y)
### Không gian RGB sang không gian HSV
Không gian màu RGB có khoảng ~16,8 triệu màu rất tốn tài nguyên để biểu diễn hết, trong khi HSV có thể biểu diễn lại chỉ cần một không gian nhỏ hơn tầm 300 ngàn màu. Đây là lí do người ta thường hay xử lý trong không gian HSV, dưới đây công thức chuyển đổi

Đại lượng H được tính bởi biểu thức sau
$$
H =
\begin{cases}
\theta & \text{if } B \leq G \\
360 - \theta & \text{if } B > G
\end{cases}
$$

Với góc $\theta$ được tính

$$
\theta = \cos^{-1} \left\{
\frac{
\frac{1}{2} \left[ (R - G) + (R - B) \right]
}{
\left[ (R - G)^2 + (R - B)(G - B) \right]^{1/2}
}
\right\}
$$

Đại lượng S được tính dựa trên

$$
S = 1 - \frac{3}{R + G + B} \left[ \min(R, G, B) \right]
$$

Cuối cùng Value được tính dựa trên

$$
V = \frac{1}{3}(R + G + B)
$$

- Chuyển từ không gian HSV sang không gian RGB

Giả sử các đại lượng đã được chuẩn hóa trong đoạn $[0,1]$, với $H \in [0,360^\circ]$.

Ta chia theo 3 miền giá trị của Hue:

**1) Sector RG ($0^\circ \leq H < 120^\circ$)**
$$
B = V(1-S)
$$
$$
R = V\left[1 + \frac{S\cos H}{\cos(60^\circ - H)}\right]
$$
$$
G = 3V - (R + B)
$$

**2) Sector GB ($120^\circ \leq H < 240^\circ$)**
Đặt $H' = H - 120^\circ$, khi đó:
$$
R = V(1-S)
$$
$$
G = V\left[1 + \frac{S\cos H'}{\cos(60^\circ - H')}\right]
$$
$$
B = 3V - (R + G)
$$

**3) Sector BR ($240^\circ \leq H \leq 360^\circ$)**
Đặt $H'' = H - 240^\circ$, khi đó:
$$
G = V(1-S)
$$
$$
B = V\left[1 + \frac{S\cos H''}{\cos(60^\circ - H'')}\right]
$$
$$
R = 3V - (G + B)
$$

Sau khi tính xong, nếu cần về thang 8-bit thì nhân mỗi kênh với 255 và ép về đoạn $[0,255]$.



### Không gian RGB sang không gian LAB


#### Bước 1: Chuẩn hóa RGB và khử gamma (sRGB)

Cho giá trị RGB ban đầu trong khoảng $[0,255]$:

$$
R' = \frac{R}{255}, \quad G' = \frac{G}{255}, \quad B' = \frac{B}{255}
$$

Khử gamma:

$$
C_{lin} =
\begin{cases}
\frac{C'}{12.92} & \text{nếu } C' \le 0.04045 \\
\left(\frac{C' + 0.055}{1.055}\right)^{2.4} & \text{nếu } C' > 0.04045
\end{cases}
$$

Áp dụng cho $R', G', B'$ để thu được:

$$
R_{lin}, \quad G_{lin}, \quad B_{lin}
$$


#### Bước 2: Chuyển từ Linear RGB → XYZ

Sử dụng ma trận chuyển đổi (chuẩn sRGB, illuminant D65):

$$
\begin{bmatrix}
X \\
Y \\
Z
\end{bmatrix}
=
\begin{bmatrix}
0.4124 & 0.3576 & 0.1805 \\
0.2126 & 0.7152 & 0.0722 \\
0.0193 & 0.1192 & 0.9505
\end{bmatrix}
\begin{bmatrix}
R_{lin} \\
G_{lin} \\
B_{lin}
\end{bmatrix}
$$


#### Bước 3: Chuẩn hóa XYZ theo điểm trắng D65

$$
X_n = 95.047, \quad Y_n = 100.000, \quad Z_n = 108.883
$$

$$
x = \frac{X}{X_n}, \quad y = \frac{Y}{Y_n}, \quad z = \frac{Z}{Z_n}
$$


#### Bước 4: Áp dụng hàm phi tuyến

$$
f(t) =
\begin{cases}
t^{1/3} & \text{nếu } t > 0.008856 \\
7.787t + \frac{16}{116} & \text{ngược lại}
\end{cases}
$$


#### Bước 5: Tính giá trị LAB

$$
L^* = 116 f(y) - 16
$$

$$
a^* = 500 \big(f(x) - f(y)\big)
$$

$$
b^* = 200 \big(f(y) - f(z)\big)
$$

$$
RGB \rightarrow RGB_{linear} \rightarrow XYZ \rightarrow LAB

### Không gian màu trong bài toán Classification và phương sai giải thích 
Trong thực tế, khi xử lý ảnh, chúng ta phải đối mặt với rất nhiều vấn đề liên quan đến tài nguyên xử lý như tập dữ liệu rất lớn (hàng ngàn, hàng vạn bức ảnh), mỗi bức ảnh thường có quality từ trung bình trở lên dẫn đến việc xử lý đơn lẻ một bức ảnh cũng có thể kiến cho máy tính xử lý phức tạp hơn. Vì thế họ mới cố gắng tìm mọi cách để đưa dữ liệu về dạng dễ sử dụng hơn, chỉ tập trung vào đặc trưng, dễ cho máy tính có thể tính toán. Một trong những cách đó chính là chuyển đổi giữa các không gian màu. Tưởng tượng một ảnh RGB có thể có tới ~16,8 triệu cách biểu diễn màu trên một pixel, nhưng nếu đưa qua không gian HSV thì chỉ cần hơn ~300.000 giá trị để có thể biểu diễn 16,8 triệu màu RGB. Nhưng nói đi cũng phải nói lại, đánh đổi việc tiết kiệm chính là đối mặt với sự mất mát thông tin. Đối với người bình thường, chỉ cần chuyển một chiếc lá từ RGB sang HSV thì có thể con người đã không thể nhận ra đó là chiếc lá. Vì vậy khi chọn không gian màu phù hợp, người ta đã đưa ra khái niệm **Phương sai giải thích** để đo đạc mức độ mất mát thông tin.

#### Phương sai giải thích
- Phương sai giải thích (Explained Variance) là một khái niệm trong thống kê và học máy, dùng đo mức độ mà mô hình hoặc một biến có thể giải thích được sự biến thiên của dữ liệu.
- Giá trị dữ liệu luôn có sự giao động mà người ta tính toán được gọi là phương sai, mô hình có thể giải thích được một phần được gọi là phương sai giải thích
- VD: Người ta mong muốn khảo sát được thời gian trung bình di chuyển từ nhà đến trường học, mô hình xây dựng dựa vào "Số đèn đỏ" mà bạn gặp để dự đoán, giải thích và cho rằng số đèn đỏ là một trong yếu tố ảnh hưởng đến thời gian đi học. Phần mà mô hình giải thích được, được gọi là phương sai giải thích.
- Liên hệ đến việc chuyển đổi giữa các không gian màu, ta sẽ dùng PCA để tìm các trục trong không gian. PCA sẽ chọn các trục có phương sai lớn nhất để giữ lại, từ đó lượng thông tin được giữ chính là phương sai được chiếu theo trục đó. Từ đó ta có công thức tính phương sai giải thích của thành phần thứ k:

### Tổng phương sai

$$
\mathrm{Var}_{total} = \sum_{i=1}^{d} \lambda_i
$$

Trong đó:
- $\lambda_i$ là eigenvalue thứ $i$
- $d$ là số chiều ban đầu của dữ liệu

##### Phương sai giải thích của thành phần thứ k

$$
\mathrm{Explained\ Variance\ Ratio}_k = \frac{\lambda_k}{\sum_{i=1}^{d} \lambda_i}
$$

##### Tổng phương sai giải thích với k thành phần

$$
\mathrm{Total\ Explained\ Variance} = \sum_{k=1}^{K} \frac{\lambda_k}{\sum_{i=1}^{d} \lambda_i}
$$

##### Dạng vector (thường dùng trong PCA)

Giả sử vector eigenvalues:

$$
\Lambda = [\lambda_1, \lambda_2, ..., \lambda_d]
$$

Thì:

$$
\mathrm{Explained\ Variance\ Ratio} = \frac{\Lambda}{\sum_{i=1}^{d} \lambda_i}
$$

## 1. Mô tả dạng thống kê kiểm định, cách thực hiện và Giả thuyết H0
**Kiểm định phân phối chuẩn** (Shapiro-Wilk, Kolmogorov-Smirnov, Anderson-Darling) là những thủ tục thống kê lý thuyết (theory-driven methods) nhằm đánh giá xem một mẫu dữ liệu quan sát độc lập có được rút ra từ một quần thể mang phân phối chuẩn hay không.
*   **Cách thực hiện:** Các kiểm định này vận hành bằng cách so sánh hàm phân phối của dữ liệu thực nghiệm với một phân phối chuẩn lý thuyết (có cùng giá trị trung bình và độ lệch chuẩn) để xem mức độ khớp nhau. Các công cụ phần mềm (như SPSS, R) sẽ tính toán ra một giá trị thống kê và giá trị p-value tương ứng.
*   **Giả thuyết H0:** Giả thuyết null ($H_0$) trong các kiểm định này luôn là: **Dữ liệu tuân theo quy luật phân phối chuẩn**. Nếu p-value < mức ý nghĩa $\alpha$ (thường là 0.05), ta bác bỏ $H_0$, tức là dữ liệu không có phân phối chuẩn. Ngược lại, nếu p-value $\ge 0.05$, ta không có đủ cơ sở bác bỏ $H_0$ và chấp nhận dữ liệu có phân phối chuẩn.

<a id="sec-2"></a> 

## 2. Mục đích của việc kiểm định phân phối chuẩn giúp ích gì cho end-user?
*   **Thông tin đạt được:** Nó cho người dùng biết dữ liệu có đáp ứng điều kiện tiên quyết về phân phối chuẩn hay không.
*   **Giúp ích như thế nào:** Toàn bộ khung phương pháp của **thống kê tham số** (parametric statistical methods) như kiểm định T-test, ANOVA, hay hồi quy tuyến tính đều dựa trên nền tảng giả định dữ liệu có phân phối chuẩn. Nếu giả định này bị vi phạm, các phân tích, diễn giải và suy diễn thống kê phía sau có thể bị sai lệch và hoàn toàn không hợp lệ. Việc kiểm định giúp người dùng biết để hoặc là tin tưởng áp dụng thống kê tham số, hoặc là chuyển sang các phương pháp **thống kê phi tham số** (non-parametric) không đòi hỏi giả định phân phối.

## 3. Bảng so sánh ưu nhược điểm, ý nghĩa và áp dụng của các loại kiểm định phổ biến
Dưới đây là so sánh 3 loại kiểm định mạnh và phổ biến nhất: Shapiro-Wilk (SW), Kolmogorov-Smirnov (KS) và Anderson-Darling (AD).

| Loại kiểm định | Ý nghĩa thống kê & Ưu điểm | Nhược điểm | Áp dụng thực tế | Ví dụ / Tình huống |
| :--- | :--- | :--- | :--- | :--- |
| **Shapiro-Wilk (SW)** | Là kiểm định **mạnh nhất (most powerful)** để phát hiện sự lệch chuẩn ở mọi loại phân phối. Đánh giá sự tương quan giữa dữ liệu và điểm chuẩn kỳ vọng. | **Rất nhạy** với mẫu lớn ($n \ge 200$), dẫn tới mắc sai lầm loại I (báo dữ liệu không chuẩn dù chỉ lệch rất nhỏ). Bị ảnh hưởng nếu dữ liệu có quá nhiều giá trị trùng lặp (ties). | Tốt nhất và ưu tiên dùng cho **mẫu nhỏ ($n \le 50$)**. | Kiểm tra mức độ chuẩn của mẫu thu nhập 30 người. SW sẽ báo kết quả rất chính xác. |
| **Kolmogorov-Smirnov (KS)** | Đo lường khoảng cách lớn nhất giữa hàm phân phối tích lũy kinh nghiệm và lý thuyết. **Ít bị sai lầm loại I ở mẫu lớn**. | Sức mạnh (power) **yếu nhất** trong các kiểm định, độ nhạy rất kém ở mẫu nhỏ. | Ưu tiên dùng cho **mẫu trung bình đến lớn ($n > 50$ hoặc $n > 60$)**. | Khảo sát điểm thi của 500 học sinh. KS sẽ tránh việc bác bỏ nhầm phân phối chuẩn do các nhiễu nhỏ. |
| **Anderson-Darling (AD)** | Là phiên bản tinh chỉnh, **đặt trọng số cao vào hai đuôi (tails)** của phân phối. Sức mạnh chỉ đứng sau SW. | Tương tự SW, AD dễ mắc sai lầm Loại I (báo sai dữ liệu chuẩn thành không chuẩn) khi cỡ mẫu lớn. | Dùng tốt cho cỡ mẫu từ **20 đến 60**, đặc biệt khi quan tâm tới các giá trị cực đoan ở đuôi. | Khảo sát khả năng chịu tải trọng của sợi thép, nơi các rủi ro đứt gãy ở 2 đuôi cực đoan cần được phát hiện nhạy bén. |

## 4. Về mốc n=5000 và các mốc quy định cỡ mẫu khác
*   **Tại sao có mốc $n=5000$:** Thuật toán gốc của Shapiro-Wilk ban đầu chỉ được thiết kế với giới hạn cỡ mẫu $n \le 50$. Về sau, Royston (năm 1995) đã đề xuất một thuật toán cải tiến (Algorithm AS R94) cho phép tính toán trọng số SW áp dụng được cho bất kỳ mẫu nào nằm trong khoảng **$3 \le n \le 5000$**. Trên 5000, thuật toán tiêu tốn quá nhiều tài nguyên máy tính, không hội tụ chính xác và các phần mềm thống kê bị giới hạn tính toán theo thuật toán này.
*   **Các mốc thực tiễn khác:** Trong thống kê ứng dụng, người ta không đợi đến $n=5000$ mới đổi kiểm định.
    *   **$n \le 50$:** SW là bắt buộc và tối ưu nhất.
    *   **$n > 60$:** KS bắt đầu vượt trội hơn vì giữ được tỷ lệ sai lầm Loại I thấp.
    *   **$n \ge 200$ (Mẫu lớn):** Các kiểm định SW và AD trở nên quá nhạy cảm, chỉ một độ lệch cực nhỏ so với phân phối chuẩn lý thuyết cũng khiến chúng báo "không chuẩn" (p < 0.05). Lúc này, lời khuyên là **không nên chỉ dựa vào p-value của kiểm định**, mà phải kết hợp xem **biểu đồ (Histogram, Q-Q Plot)** và các hệ số **Skewness/Kurtosis** để ra quyết định.

## 5. Trong thực tế người ta hay dùng 2 loại kiểm định này để làm gì?
Trong phân tích thực tế, SW và KS thường được dùng ở bước **tiền xử lý và kiểm tra giả định (assumption checking)** để chọn hướng phân tích dữ liệu:
1.  Nếu kết quả là **chuẩn**: Nhà nghiên cứu được phép sử dụng các **thống kê tham số** mạnh mẽ (như Paired T-test, Independent T-test, One-way ANOVA).
2.  Nếu kết quả là **không chuẩn**: Nhà nghiên cứu biết rằng áp dụng thống kê tham số sẽ sai lệch, do đó họ phải chọn các phương pháp **thống kê phi tham số** tương ứng (như Wilcoxon rank-sum, Kruskal-Wallis).

## 6. Chuẩn hóa dữ liệu phân phối chuẩn và không chuẩn
*   **Nếu dữ liệu đã có phân phối chuẩn:** Về mặt hình thái phân phối, dữ liệu đã đáp ứng yêu cầu nên thường không cần phải chuẩn hóa thêm về hình thái nữa. *.*
*   **Nếu dữ liệu không có phân phối chuẩn:** Ta dùng các phương pháp để biến đổi chúng nhằm đưa về phân phối chuẩn, bao gồm các phép toán như **lấy Logarit (Log(X)), biến đổi Arcsin, hoặc phép biến đổi tổng quát Box-Cox**.
*   **Tại sao phải chuẩn hóa?** Bởi vì các phương pháp thống kê tham số thường có sức mạnh kiểm định (power) cao hơn, đưa ra các kết luận sâu sắc hơn về mặt định lượng so với phi tham số. Nếu thông qua các phép chuẩn hóa (biến đổi) mà dữ liệu trở thành chuẩn, ta có thể tiếp tục sử dụng các kiểm định tham số truyền thống. Tuy nhiên, nếu dùng phép biến đổi (như Log, Box-Cox) mà vẫn không đưa được về chuẩn, ta buộc phải chấp nhận từ bỏ thống kê tham số để dùng phương pháp phi tham số (non-parametric).

## 7. Thực hành: Chạy kiểm định trên dữ liệu với Shapiro & Pearson

```python
from table.dataset import TableDataset
from config.settings import PATH_FOLDER_TABLE_RAW
from statistic.test_distribution import ShapiroWilkTesting, AgostinoPearsonTesting
from pathlib import Path
dataset = TableDataset(str(Path(PATH_FOLDER_TABLE_RAW) / "Building_Permits.csv"))
dataset.load()
dataset.info()
```

```python
# 1. Chạy class Shapiro
shapiro_tester = ShapiroWilkTesting(column_name="Estimated Cost", alpha=0.05)
shapiro_tester._dataset = dataset
shapiro_tester.run()
shapiro_tester.log()

# 2. Chạy class Pearson (Chỉ cần thay tên class)
pearson_tester = AgostinoPearsonTesting(column_name="Estimated Cost", alpha=0.05)
pearson_tester._dataset = dataset
pearson_tester.run()
pearson_tester.log()
```

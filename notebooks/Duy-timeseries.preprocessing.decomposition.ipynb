#### Tại sao lại phải phân rã chuỗi và hệ quả nếu không thực hiện?
Việc phân rã chuỗi thời gian áp dụng tư tưởng chia một bài toán phức tạp thành các phần nhỏ hơn, dễ hiểu hơn. Dữ liệu chuỗi thời gian thường chứa các mô hình đan xen nhau; do đó, việc phân rã giúp bóc tách chuỗi thành các thành phần riêng biệt: **xu hướng (trend-cycle)**, **tính mùa vụ (seasonal)**, và **phần dư/nhiễu (irregular/remainder)**. Mục đích của việc làm phẳng (smoothing) này là loại bỏ tính ngẫu nhiên, giúp ta hiểu rõ hành vi của dữ liệu để dự báo tương lai với độ chính xác cao hơn. 
* **Xu hướng (Trend):** Chiều hướng tăng, giảm hoặc đi ngang trong dài hạn (Ví dụ: Sự hình thành của một làn sóng lây nhiễm COVID-19 mới).
* **Tính thời vụ (Seasonal):** Sự dao động lặp đi lặp lại có tính chu kỳ cố định (Ví dụ: Số ca báo cáo luôn suy giảm vào các ngày cuối tuần do nhân viên y tế nghỉ ngơi, và tăng vọt trở lại vào đầu tuần do cộng dồn số liệu).
* **Phần dư / Nhiễu ngẫu nhiên (Residual/Irregular):** Những biến động dị thường không lường trước được, không thuộc về xu hướng hay chu kỳ thời vụ.
**Hệ quả nếu không phân rã:** Nếu để nguyên chuỗi dữ liệu gốc mà không tách mùa vụ và xu hướng, các thông tin quan trọng sẽ bị che lấp, dẫn đến nhận định sai lầm.
*   **Ví dụ thực tế:** Giả sử vào một ngày Chủ nhật, có một ổ dịch mới bùng phát khiến số ca thực tế tăng bất thường thêm 500 ca. Tuy nhiên, vì đặc thù mùa vụ của ca nhiễm vào "cuối tuần", số liệu báo cáo ngày Chủ nhật luôn thấp hơn ngày thường khoảng 1,000 ca. Nếu chỉ nhìn vào biểu đồ gốc, tổng số ca ngày Chủ nhật đó vẫn thấp hơn thứ Tư tuần trước, khiến hệ thống y tế hoàn toàn bỏ sót điểm bùng phát (outlier) này. Nhưng nếu ta đã bóc tách và deseasonalized, điểm bất thường kia sẽ ngay lập tức hiện rõ trên phần Nhiễu (Residual).

#### Ký hiệu trong công thức của mô hình
Trong các mô hình phân rã, chuỗi thời gian được biểu diễn qua các biến số sau:
* **$t$ (Time):** Đại diện cho một mốc thời gian cụ thể (ví dụ: ngày thứ $t$). Khi đứng ở dưới dạng chỉ số phụ, nó ám chỉ giá trị của biến đang được xét tại thời điểm đó.
* **$Y_t$ (Yield / Observed Value):** Giá trị quan sát thực tế thu thập được tại thời điểm $t$. Ví dụ, $Y_t$ chính là số ca mắc COVID-19 mới được công bố vào ngày $t$.
* **$T_t$ (Trend):** Thành phần Xu hướng tại thời điểm $t$. Thể hiện trạng thái vĩ mô của dịch bệnh đang bùng phát (dốc lên) hay đang được kiểm soát (dốc xuống).
* **$S_t$ (Seasonal):** Thành phần Thời vụ tại thời điểm $t$. Thể hiện sự tác động của chu kỳ (như hiệu ứng ngày nghỉ cuối tuần) lên dữ liệu.
* **$R_t$ hoặc $E_t$ (Residual / Error):** Phần dư, sai số ngẫu nhiên tại thời điểm $t$. Đây là độ lệch còn sót lại sau khi đã loại bỏ Xu hướng và Tính thời vụ khỏi dữ liệu gốc.

#### Phân rã Additive (Mô hình Cộng)
* **Công thức:** $Y_t = T_t + S_t + R_t$
* **Ý nghĩa:** Cường độ dao động của yếu tố thời vụ là một **hằng số cố định về mặt trị số tuyệt đối**, không bị ảnh hưởng bởi việc xu hướng chung đang ở mức cao hay thấp.
* **Ví dụ với COVID-19:** Giả sử hiệu ứng cuối tuần luôn làm giảm báo cáo đi một lượng cố định là 200 ca.
    * Lúc dịch êm đềm (Xu hướng 1,000 ca/ngày): Ngày thường báo cáo 1,000 ca, cuối tuần báo cáo $1000 - 200 = 800$ ca.
    * Lúc đỉnh dịch (Xu hướng 10,000 ca/ngày): Ngày thường 10,000 ca, cuối tuần báo cáo $10000 - 200 = 9,800$ ca. Khoảng cách chênh lệch luôn là 200 ca. Mô hình này chỉ phù hợp nếu năng lực xét nghiệm của quốc gia đó có một giới hạn trần cố định không đổi.

#### Phân rã Multiplicative (Mô hình Nhân)
* **Công thức:** $Y_t = T_t \times S_t \times R_t$
* **Ý nghĩa:** Cường độ dao động của yếu tố thời vụ là một **tỷ lệ phần trăm (%)** hoặc hệ số. Biên độ dao động của nó sẽ mở rộng (nở ra) hoặc thu hẹp (co lại) tỷ lệ thuận với sự thay đổi của xu hướng.
* **Ví dụ với COVID-19:** Giả sử cuối tuần lượng xét nghiệm luôn sụt giảm 20% (hệ số thời vụ $S_t = 0.8$).
    * Lúc dịch êm đềm (Xu hướng 1,000 ca): Cuối tuần báo cáo $1000 \times 0.8 = 800$ ca (độ lệch là 200 ca).
    * Lúc đỉnh dịch bùng nổ (Xu hướng 10,000 ca): Cuối tuần báo cáo $10000 \times 0.8 = 8,000$ ca (độ lệch nới rộng lên tới 2,000 ca). Đây chính là hành vi thực tế chuẩn xác nhất của các chuỗi lây nhiễm dịch bệnh, vì khi số lượng người bệnh tăng theo cấp số nhân, sự hao hụt trong khâu báo cáo cũng tăng theo một tỷ lệ tương ứng.


#### Bộ dữ liệu phù hợp cho từng loại phân rã
*   **Dùng Additive:** Khi biên độ dao động của tính mùa vụ và phần dư giữ nguyên hoặc thay đổi không đáng kể theo thời gian, bất kể xu hướng chung đang tăng hay giảm. 
*   **Dùng Multiplicative:** Khi sự biến động của yếu tố mùa vụ tăng lên hoặc giảm đi tương ứng với sự gia tăng hoặc sụt giảm của thành phần xu hướng. Phương pháp này rất phổ biến khi phân tích các chuỗi dữ liệu tài chính, kinh tế hoặc tăng trưởng người dùng, nơi phần trăm thay đổi (tỷ lệ) phản ánh thực tế tốt hơn sự thay đổi về mặt trị số tuyệt đối.

#### Có thể biến đổi từ Multiplicative sang Additive không?
ta hoàn toàn có thể biến đổi từ Multiplicative sang Additive bằng cách sử dụng **hàm Logarit (Log)**. Khi lấy Log của chuỗi thời gian, mối quan hệ nhân sẽ trở thành mối quan hệ cộng: $\log(Y_t) = \log(S_t) + \log(T_t) + \log(E_t)$. Ngoài ra, ta cũng có thể dùng phép biến đổi Box-Cox (với $\lambda = 0$ cho mô hình nhân và $\lambda = 1$ cho mô hình cộng) để linh hoạt chuyển đổi giữa hai dạng này.

#### STL Decomposition là gì? Sự khác biệt và vai trò của LOESS 
*   **STL là gì:** STL là viết tắt của "Seasonal and Trend decomposition using Loess". Phương pháp này mạnh mẽ, linh hoạt và chỉ cung cấp phân rã dạng Additive. 
*   **Vai trò của LOESS:** LOESS (Locally Estimated Scatterplot Smoothing) là thuật toán hồi quy đa thức cục bộ. Điểm mạnh của nó là không cần phải chỉ định một hàm toán học tổng thể cứng nhắc. LOESS bóc tách dữ liệu bằng cách khớp các mô hình đơn giản vào từng tập con dữ liệu cục bộ, mang lại tính linh hoạt cao trong việc lập mô hình các chuỗi thời gian có cấu trúc phức tạp phi tuyến tính mà các phương pháp truyền thống khó nắm bắt.

| Tiêu chí | Additive / Multiplicative Cổ điển | Phương pháp STL |
| :--- | :--- | :--- |
| **Tính linh hoạt của Yếu tố thời vụ** | Tĩnh (Static). Tính thời vụ bị ép buộc lặp lại y hệt nhau qua mọi chu kỳ. | **Động (Dynamic).** Tính thời vụ có thể tiến hóa theo thời gian (Ví dụ: Thói quen xét nghiệm đầu năm 2020 khác hẳn cuối năm 2021 do sự xuất hiện của vaccine). |
| **Khả năng kháng nhiễu (Robustness)** | Rất kém. Một ngày dồn số liệu báo cáo đột biến (Outlier) sẽ làm sai lệch toàn bộ đường Xu hướng. | **Cực tốt.** Thuật toán tự động cô lập các cú sốc (Outliers) và đẩy chúng thẳng vào phần Nhiễu ($R_t$), giúp đường Xu hướng duy trì được sự ổn định. |
| **Xử lý chu kỳ** | Chỉ hoạt động tốt với các chu kỳ đơn giản. | Xử lý linh hoạt mọi tần suất chu kỳ (giờ, ngày, tuần, tháng). |
| **Thuật toán cốt lõi** | Tính Trung bình trượt (Moving Average) cơ bản. | Sử dụng **LOESS** (Hồi quy đa thức cục bộ – Locally Estimated Scatterplot Smoothing). |

#### Tầm quan trọng của Tham số trong STL\
Khác với phương pháp cổ điển, STL cho phép Data Scientist tinh chỉnh mô hình thông qua các cửa sổ thời gian (Windows).

* **Chọn Period (Chu kỳ) định danh:** Với dữ liệu COVID-19 theo ngày, chu kỳ chuẩn xác nhất là **Period = 7**. Lý do là virus lây lan không có ngày nghỉ, nhưng hệ thống thống kê y tế của con người lại vận hành theo chu kỳ làm việc 7 ngày/tuần.
* **Seasonal Window (Cửa sổ thời vụ):** Kiểm soát tốc độ biến thiên của yếu tố thời vụ. Đặt càng nhỏ, thành phần thời vụ biến đổi càng nhanh để thích ứng với dữ liệu mới.
* **Trend Window (Cửa sổ xu hướng):** Kiểm soát độ mượt của xu hướng. Trong công tác chống dịch, nếu chính phủ ban bố lệnh "Phong tỏa toàn diện", xu hướng số ca sẽ gãy gập rất nhanh. Nếu chọn Trend Window quá lớn (quá mượt), mô hình sẽ phản ứng chậm và đẩy sự biến động thực tế này vào phần dư (Residual). Ngược lại, khi thiết lập Trend Window đủ nhỏ, đường xu hướng sẽ "ôm sát" sự sụt giảm tức thời do hiệu quả của lệnh phong tỏa mang lại.

```python
from time_series.preprocessing_decomposition import STLDecomposition, MultiplicativeDecomposition, AdditiveDecomposition
from time_series.dataset import TimeSeriesDataset
# 1. NẠP DỮ LIỆU
# TimeSeriesDataset sẽ tự động ép kiểu Datetime và sắp xếp từ quá khứ đến hiện tại
covid_dataset = TimeSeriesDataset(path="../data/time_series/time-series-19-covid-combined.csv", time_column="Date")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


def get_daily_series(dataset, country, target_col):
    """Hàm trích xuất ca mắc mới/tử vong mới/hồi phục mới theo ngày"""
    df = dataset.data
    time_col = dataset._time_column
    
    # Gom nhóm theo ngày
    df_country = df[df['Country/Region'] == country].groupby(time_col)[target_col].sum().reset_index()
    df_country = df_country.sort_values(time_col).set_index(time_col)
    
    # Lấy diff() để ra số ca mới mỗi ngày
    daily_series = df_country[target_col].diff().dropna()
    daily_series[daily_series < 0] = 0 # Khử nhiễu âm do đính chính số liệu
    
    return daily_series

def evaluate_metrics(series, title_prefix=""):
    """Hàm quét qua 3 chu kỳ và 3 phương pháp để tìm ra model ít nhiễu nhất"""
    periods = [7, 14, 30]
    print(f"{'='*65}")
    print(f"📊 ĐÁNH GIÁ: {title_prefix}")
    print(f"{'='*65}")
    print(f"{'Phương pháp':<15} | {'Chu kỳ':<10} | {'Tỉ lệ nhiễu (Variance Ratio)':<20}")
    print("-" * 65)
    
    best_method, best_period, min_noise = "", 0, float('inf')
    
    for p in periods:
        models = {
            "Additive": AdditiveDecomposition(series, p),
            "Multiplic.": MultiplicativeDecomposition(series, p),
            "STL": STLDecomposition(series, p)
        }
        
        for name, model in models.items():
            model.run()
            noise = model.residual_variance_ratio()
            print(f"{name:<15} | {p:<10} | {noise:.4f} ({noise*100:.2f}%)")
            
            if noise < min_noise:
                min_noise, best_method, best_period = noise, name, p
                
    print("-" * 65)
    print(f"🏆 TỐI ƯU: Phương pháp {best_method.upper()} với chu kỳ {best_period} ngày (Nhiễu: {min_noise*100:.2f}%)\n")
    return best_method, best_period
```

```python
# Lấy dữ liệu Ca nhiễm (Confirmed) của Mỹ (US) làm đại diện chuẩn
series_us_confirmed = get_daily_series(covid_dataset, 'US', 'Confirmed')

print(" PHÂN TÍCH PHƯƠNG PHÁP VÀ CHU KỲ TỐI ƯU")
evaluate_metrics(series_us_confirmed, "US - Daily Confirmed Cases")

# Dựa vào kết quả in ra, bạn sẽ thấy Chu kỳ 7 ngày luôn cho nhiễu thấp nhất 
# (do chu kỳ xét nghiệm/nhập liệu cuối tuần).
```

```python
# Trích xuất riêng một làn sóng dịch bùng nổ mạnh (VD: Cuối năm 2020 tại US)
wave_series = series_us_confirmed.loc['2020-10-01':'2021-02-01']

add_model = AdditiveDecomposition(wave_series, 7)
add_model.run()

mult_model = MultiplicativeDecomposition(wave_series, 7)
mult_model.run()

print(f"Tỉ lệ nhiễu Additive      : {add_model.residual_variance_ratio()*100:.2f}%")
print(f"Tỉ lệ nhiễu Multiplicative: {mult_model.residual_variance_ratio()*100:.2f}%")

# Vẽ biểu đồ Phần dư (Residual) của 2 phương pháp để so sánh trực quan
fig, ax = plt.subplots(2, 1, figsize=(10, 6), sharex=True)
ax[0].plot(add_model.resid.index, add_model.resid, color='red', label='Additive Residual')
ax[0].legend()
ax[0].set_title("Nhiễu của Additive (Rất lớn khi đỉnh dịch)")

ax[1].plot(mult_model.resid.index, mult_model.resid, color='blue', label='Multiplicative Residual')
ax[1].legend()
ax[1].set_title("Nhiễu của Multiplicative (Ổn định hơn nhờ tính theo %)")
plt.tight_layout()
plt.show()

# Bạn dùng biểu đồ này dán vào báo cáo để chứng minh Multiplicative ôm sát 
# biên độ dao động tốt hơn Additive khi đường Trend vọt lên quá cao.
```

```python
print("SO SÁNH GIỮA CÁC KHU VỰC (US vs Hàn Quốc)")

# Lấy thêm dữ liệu của Hàn Quốc (South Korea)
series_kr_confirmed = get_daily_series(covid_dataset, 'Korea, South', 'Confirmed')

evaluate_metrics(series_us_confirmed, "KHU VỰC: US ")
evaluate_metrics(series_kr_confirmed, "KHU VỰC: South Korea ")

```

```python
series_us_deaths = get_daily_series(covid_dataset, 'US', 'Deaths')
series_us_recovered = get_daily_series(covid_dataset, 'US', 'Recovered')

# Confirmed đã đánh giá ở trên, giờ đánh giá Deaths và Recovered
evaluate_metrics(series_us_deaths, "CHỈ SỐ: US Daily Deaths (Tử vong)")
evaluate_metrics(series_us_recovered, "CHỈ SỐ: US Daily Recovered (Hồi phục)")

# Dựa vào kết quả:
# - Confirmed: Chịu ảnh hưởng chu kỳ 7 ngày mạnh nhất.
# - Deaths: Thường trễ hơn và tản mạn hơn (chu kỳ có thể là 14 ngày hoặc biến động rải rác).
# - Recovered: Rất hay bị "xả số liệu dồn", nhiễu cực cao, các phương pháp cổ điển thường thất bại.
```

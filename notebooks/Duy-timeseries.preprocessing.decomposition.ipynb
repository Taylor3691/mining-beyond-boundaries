**Tại sao lại phải phân rã chuỗi và hệ quả nếu không thực hiện?**
Việc phân rã (decomposition) chuỗi thời gian áp dụng tư tưởng chia một bài toán phức tạp thành các phần nhỏ hơn, dễ hiểu hơn. Dữ liệu chuỗi thời gian thường chứa các mô hình (pattern) đan xen nhau; do đó, việc phân rã giúp bóc tách chuỗi thành các thành phần riêng biệt: **xu hướng (trend-cycle)**, **tính mùa vụ (seasonal)**, và **phần dư/nhiễu (irregular/remainder)**. Mục đích của việc làm phẳng (smoothing) này là loại bỏ tính ngẫu nhiên, giúp ta hiểu rõ hành vi của dữ liệu để dự báo tương lai với độ chính xác cao hơn. 
**Hệ quả nếu không phân rã:** Nếu ta không tách rời các yếu tố này, sự kết hợp của chúng có thể làm lu mờ các thông tin quan trọng. Ví dụ, một điểm bất thường (anomaly) cục bộ xảy ra trong "khoảng thời gian thấp điểm" có thể vẫn có giá trị thấp hơn mức bình thường của "khoảng thời gian cao điểm", dẫn đến việc ta hoàn toàn bỏ sót điểm bất thường này nếu không trừ đi yếu tố mùa vụ. Ngoài ra, để phân tích chính xác trạng thái thực tế của một chỉ số kinh tế, ta buộc phải loại bỏ biến động mùa vụ thông qua phân rã.

**Công thức phân rã Additive (Cộng) và ý tưởng**
*   **Công thức:** $Y_t = S_t + T_t + E_t$ (hoặc $R_t$).
*   **Ý tưởng:** Mô hình cho rằng các thành phần xu hướng, mùa vụ và nhiễu cộng dồn lại với nhau để tạo thành dữ liệu chuỗi thời gian quan sát được. Cường độ biến động của tính mùa vụ hoặc nhiễu được giả định là có mức độ ổn định, không thay đổi theo độ lớn của chuỗi thời gian.

**Công thức phân rã Multiplicative (Nhân) và ý nghĩa khác biệt**
*   **Công thức:** $Y_t = S_t \times T_t \times E_t$.
*   **Sự khác biệt về ý nghĩa:** Thay vì cộng, các thành phần được nhân với nhau. Trong Additive, độ lớn của biến động mùa vụ là hằng số. Nhưng trong Multiplicative, độ lớn của biến động mùa vụ **tỷ lệ thuận (scale)** với mức độ/xu hướng của chuỗi thời gian. Nếu xu hướng tăng theo cấp số nhân, biên độ tuyệt đối của biến động mùa vụ cũng sẽ mở rộng theo.

**Bộ dữ liệu phù hợp cho từng loại phân rã**
*   **Dùng Additive:** Khi biên độ dao động của tính mùa vụ và phần dư giữ nguyên hoặc thay đổi không đáng kể theo thời gian, bất kể xu hướng chung đang tăng hay giảm. 
*   **Dùng Multiplicative:** Khi sự biến động của yếu tố mùa vụ tăng lên hoặc giảm đi tương ứng với sự gia tăng hoặc sụt giảm của thành phần xu hướng. Phương pháp này rất phổ biến khi phân tích các chuỗi dữ liệu tài chính, kinh tế hoặc tăng trưởng người dùng, nơi phần trăm thay đổi (tỷ lệ) phản ánh thực tế tốt hơn sự thay đổi về mặt trị số tuyệt đối.

**Có thể biến đổi từ Multiplicative sang Additive không?**
ta hoàn toàn có thể biến đổi từ Multiplicative sang Additive bằng cách sử dụng **hàm Logarit (Log)**. Khi lấy Log của chuỗi thời gian, mối quan hệ nhân sẽ trở thành mối quan hệ cộng: $\log(Y_t) = \log(S_t) + \log(T_t) + \log(E_t)$. Ngoài ra, ta cũng có thể dùng phép biến đổi Box-Cox (với $\lambda = 0$ cho mô hình nhân và $\lambda = 1$ cho mô hình cộng) để linh hoạt chuyển đổi giữa hai dạng này.

**STL Decomposition là gì? Sự khác biệt và vai trò của LOESS**
*   **STL là gì:** STL là viết tắt của "Seasonal and Trend decomposition using Loess". Phương pháp này mạnh mẽ, linh hoạt và chỉ cung cấp phân rã dạng Additive. 
*   **Khác biệt so với các loại khác:** Không giống như các kỹ thuật cổ điển hay X-11/SEATS (thường chỉ xử lý dữ liệu theo tháng hoặc quý), STL có thể xử lý **bất kỳ loại tính mùa vụ nào** (theo giờ, ngày, tuần, v.v.), cho phép tính mùa vụ biến đổi theo thời gian, kiểm soát được độ mượt của xu hướng và có khả năng **chống chịu rất tốt trước các điểm dị thường (outliers)**. 
*   **Vai trò của LOESS:** LOESS (Locally Estimated Scatterplot Smoothing) là thuật toán hồi quy đa thức cục bộ. Điểm mạnh của nó là không cần phải chỉ định một hàm toán học tổng thể cứng nhắc. LOESS bóc tách dữ liệu bằng cách khớp các mô hình đơn giản vào từng tập con dữ liệu cục bộ, mang lại tính linh hoạt cao trong việc lập mô hình các chuỗi thời gian có cấu trúc phức tạp phi tuyến tính mà các phương pháp truyền thống khó nắm bắt.

**Sự quan trọng của Period (Chu kỳ) cho STL và cách chọn cho chuẩn**
*   **Mức độ quan trọng:** Độ dài của chu kỳ ($T$) là tham số bản lề để xác định các mô hình lặp lại. Nếu ta xác định sai khoảng chu kỳ (đặc biệt đối với dữ liệu IoT lớn với độ dài chu kỳ dài như 1440 nếu thu thập từng phút), thuật toán sẽ thất bại trong việc trích xuất tính mùa vụ thực sự.
*   **Cách chọn tham số cho chuẩn:**
    Trong thuật toán STL, việc lựa chọn xoay quanh 2 tham số chính (cửa sổ): **trend window** và **seasonal window**. Cả hai cửa sổ này đều **bắt buộc phải là số lẻ**.
    *   *Seasonal window* kiểm soát tốc độ thay đổi của tính mùa vụ theo thời gian. Nếu đặt tham số này càng nhỏ, cấu trúc mùa vụ càng linh hoạt. Việc thiết lập nó ở mức vô hạn (bằng cách đặt *window = 'periodic'*) sẽ bắt buộc thành phần mùa vụ lặp lại y hệt qua các năm. 
    *   *Trend window* kiểm soát độ mượt (wiggliness) của xu hướng. Nếu ta thấy tín hiệu dữ liệu từ những đợt thay đổi mạnh (như cuộc khủng hoảng tài chính) vô tình bị lọt vào phần dư (remainder), chứng tỏ cửa sổ xu hướng đang quá dài, làm hệ thống quá cứng nhắc. Để chuẩn nhất, ta cần **chọn trend window ngắn hơn** để xu hướng phản ứng linh hoạt với các thay đổi nhanh chóng của dữ liệu. Trong nhiều cấu hình tự động, STL mặc định gán season window là 11 hoặc 13 và tự động tính toán trend window, nhưng ta luôn cần kiểm tra và tinh chỉnh lại bằng tay nếu cần.

# Lý thuyết về Class Imbalance

## 1. Thế nào là mất cân bằng lớp?
- Để hiểu về mất cân bằng lớp, cần nhìn lại lý thuyết về các độ đo để đánh giá hiệu suất cho bài toán phân lớp (Classification). Thông thường, độ đo phổ biến nhất sẽ là Accuracy – tỉ lệ các bản ghi (record) được dự đoán đúng trên tổng số bản ghi. Tuy nhiên, Accuracy chỉ thực sự có ý nghĩa khi phân phối giữa các lớp là tương đối đồng đều.

- Từ cơ sở đó, khái niệm mất cân bằng lớp (Class Imbalance) được sinh ra để thể hiện tình trạng phân phối dữ liệu không đồng đều, trong đó lớp mà chúng ta quan tâm nhất (Positive Class) lại có tần suất xuất hiện cực thấp so với lớp còn lại (negative class). Trong trường hợp này, việc sử dụng độ đo Accuracy sẽ dẫn đến kết quả sai lệch vì nó bị lấn át bởi hiệu suất của lớp đa số. Để có thể đánh giá đúng về mặt hiệu suất, có thể xem xét thêm ở hai độ đo khác là:
    + Sensitivity hay Recall: Khả năng nhận diện đúng các bản ghi thuộc Positive Class. Công thức sẽ là:$$Sensitivity = \frac{TP}{P}$$
    + Specificity: Khả năng nhận diện đúng các bản ghi thuộc Negative Class. Công thức sẽ là $$Specificity = \frac{TN}{N}$$

- Mất cân bằng lớp cũng chính là hiện tượng khi một bộ phận lớp có thể đạt chỉ số Accuracy và Specificity rất cao nhưng lại có Sensitivity cực thấp, khiến nó trở nên vô dụng trong việc giải quyết các bài toán tìm kiếm các trường hợp hiếm.

## 2. Làm sao để xác định mất cân bằng?
- Việc xác định một tập dữ liệu có bị mất cân bằng hay không không chỉ dựa vào một con số phần trăm nhất định (như dưới 5% hay 10%), mà dựa trên sự kết hợp giữa phân phối thống kê và mục tiêu của bài toán phân lớp.

- Đầu tiên là ở Accuracy: Dấu hiệu đầu tiên để xác định mất cân bằng là khi chỉ số Accuracy trở nên quá tốt để có thể tin tưởng vào trong việc nhận định nhưng mô hình lại thất bại trong việc tìm ra lớp mục tiêu. Nếu mô hình có xu hướng dự đoán mọi thứ thuộc về lớp đa số để tối ưu Accuracy, đó chính là trạng thái mất cân bằng. Nội dung này dựa trên trích dẫn từ sách "Data Mining Concepts and Technique 4th , Digital Image Processing 3rd" (trang 367) như sau: "An accuracy rate of, say, 97% may make the classifier seem quite accurate, but what if only, say, 3% of the training tuples are actually cancer?". 

- Ở Sensitivity: Một dấu hiệu nữa là ở chỉ số Sensitivity (khả năng nhận diện Positive Class) sụt giảm nghiêm trọng so với Specificity (khả năng nhận diện Negative Class). Trích dẫn từ sách "Data Mining Concepts and Technique 4th , Digital Image Processing 3rd" (trang 367) có nhấn mạnh việc cần các thước đo đánh giá riêng biệt cho từng lớp như sau: "...we need other measures, which access how well the classifier can recognize the positive tuples and how well it can recognize the negative tuples."

- Ngoài ra còn có tỷ lệ mất cân bằng (Imbalance Ratio - IR) là thước đo định lượng để đánh giá mức độ nghiêm trọng của sự chênh lệch giữa các lớp. Cách tính toán là:$$IR = \frac{\text{Số lượng bản ghi của lớp đa số (Majority Class)}}{\text{Số lượng bản ghi của lớp thiểu số (Minority Class)}}$$Phân loại mức độ mất cân bằng như sau:

  + Mất cân bằng nhẹ (Mild): Khi IR dao động nhỏ (ví dụ từ 1:1 đến 9:1).
  + Mất cân bằng trung bình (Moderate): Khi IR từ 10:1 đến 100:1. Tại đây, ranh giới phân lớp bắt đầu bị lấn át bởi lớp đa số.
  + Mất cân bằng nghiêm trọng (Extreme): Khi IR > 100:1 . Tại mức này, lớp thiểu số có thể bị coi là nhiễu (noise) và bị mô hình bỏ qua hoàn toàn.

## 3. Mất cân bằng lớp gây ảnh hưởng gì?
- Ảnh hưởng của mất cân bằng lớp nằm ở:
    + Sự thiên kiến (bias) của thuật toán đối với lớp đa số khi trong điều kiện mất cân bằng, các mô hình truyền thống có xu hướng ưu tiên việc dự đoán đúng các bản ghi thuộc lớp đa số để đạt được Accuracy cao nhất. Kết quả là mô hình vô tình bỏ qua lớp thiểu số và coi chúng là những trường hợp không đáng kể. Nếu mục tiêu bài toán nhắm vào lớp thiểu số này thì coi như mô hình phân loại trở nên vô dụng.
    + Mất cân bằng lớp còn làm cho chỉ số Accuracy trở nên vô nghĩa khi mà mô hình có thể đạt độ chính xác 99.9% bằng cách luôn dự đoán mọi thứ cho là thông thường, nhưng lại thất bại hoàn toàn trong việc tìm ra 0.1% trường hợp hiếm mà chúng ta thực sự quan tâm.
    + Ảnh hưởng nghiêm trọng nhất là sự chênh lệch về chi phí khi dự đoán sai. Trong các bài toán đi kèm với dữ liệu mất cân bằng, việc dự đoán sai một bản ghi của lớp xuất hiện ở tần suất hiếm (False Negative) thường để lại hậu quả nặng nề hơn rất nhiều so với việc dự đoán sai một bản ghi của lớp xuất hiện đa số trong dữ liệu (False Positive). Dù vậy thì các thuật toán học máy truyền thống lại coi hai loại lỗi này có giá trị ngang nhau.
- Để làm rõ sự ảnh hưởng này, có thể đi qua một ví dụ như sau về chẩn đoán bệnh Ung thư (đây là một bệnh có tỉ lệ hiếm):
    + Lấy bối cảnh là khi giả sử trong 1 tập dữ liệu y tế có 10000 bản ghi, trong đó chỉ có 100 người thực sự bị ung thư (Positive Class - 1%) và 9900 người khỏe mạnh (Negative Class - 99%).
    + Ảnh hưởng của việc mất cân bằng lớp xảy ra khi nếu mô hình dự đoán tất cả 10000 người là khỏe mạnh, nó vẫn đạt Accuracy là 99%. Tuy nhiên, mô hình này hoàn toàn vô dụng vì nó đã bỏ sót toàn bộ 100 người bị bệnh.
    + Hậu quả về mặt chi phí là việc chẩn đoán nhầm một người có bệnh thành khỏe mạnh (False Negative) có thể dẫn đến tử vong cho bệnh nhân, và đây thực sự vô cùng nghiêm trọng so với việc chẩn đoán nhầm người khỏe mạnh thành có bệnh (False Positive) chỉ dẫn đến việc thực hiện thêm các xét nghiệm bổ sung và không để lại ảnh hưởng quá lớn. Chi phí của hai loại lỗi này là hoàn toàn khác nhau.
